```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error

df=pd.read_csv('housing.csv')
x=df[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms' , 'Area Population']]
y=df['Price']
xtrain , xtest,ytrain,ytest=train_test_split(x,y,test_size=0.25,random_state=10)
model=LinearRegression()
model.fit(xtrain,ytrain)
model.coef_
pd.DataFrame(model.coef_,columns=['coef_'],index=x.columns)
ypred=model.predict(xtest)
final_model=model
pd.to_pickle(final_model,'HousePricePredictor')
model=pd.read_pickle('HousePricePredictor')

import tkinter as tk
app=tk.Tk()
app.geometry("400x250")
app.configure(bg='#34568B')
app.title('House price predictor')

income=tk.StringVar()
house_age=tk.StringVar()
number_of_rooms=tk.StringVar()
population=tk.StringVar()

tk.Label(app,text="INCOME",bg='#34568B').place(x=20,y=30)
tk.Label(app,text="HOUSE AGE",bg='#34568B').place(x=20,y=60)
tk.Label(app,text="NUMBER OF ROOMS",bg='#34568B').place(x=20,y=90)
tk.Label(app,text="POPULATION",bg='#34568B').place(x=20,y=120)

tk.Entry(app,width=30,bg="#ffffff",textvariable=income).place(x=150,y=30)
tk.Entry(app,width=30,bg="#ffffff",textvariable=house_age).place(x=150,y=60)
tk.Entry(app,width=30,bg="#ffffff",textvariable=number_of_rooms).place(x=150,y=90)
tk.Entry(app,width=30,bg="#ffffff",textvariable=population).place(x=150,y=120)


def Predict_the_price():
    i=income.get()
    h=house_age.get()
    n=number_of_rooms.get()
    p=population.get()
    income.set('')
    house_age.set('')
    number_of_rooms.set('')
    population.set('')
    query=pd.DataFrame({'Avg. Area Income':[i],
                    'Avg. Area House Age':[h],
                    'Avg. Area Number of Rooms':[n],
                    'Area Population':[p]
                   })
    
    
    
    Predict_the_price_var=tk.Variable(app)
    Predict_the_price_var.set('%.1f$'%(model.predict(query)[0]))
    
    
    Predict_the_price=tk.Label(app,text='price for the house can be :',bg='#34568B',font=('Bahnschrift SemiBold', 10, 'bold')).place(x=30,y=200)
    Predict_the_price=tk.Label(app,text ="price for the house can be :",textvariable=Predict_the_price_var,bg='#34568B',font=('Bahnschrift SemiBold', 10, 'bold')).place(x=200,y=200)
    
tk.Button(app,text='Predict the price',command=Predict_the_price).place(x=150,y=150)

app.mainloop()
```
